### Factor Returns - Annual Rolling Returns

```python
factor_returns = (factor_data/factor_data.shift(365))-1
factor_returns.dropna(how = "all",inplace=True)
factor_returns.head()
```

```python
factor_returns.describe()
```

```python
risk_adjusted = factor_returns.mean()/factor_returns.std()
risk_adjusted = pd.DataFrame(risk_adjusted)
risk_adjusted.columns = ['Risk Adjusted']
risk_adjusted.style.background_gradient(cmap='PuBu')
```

```python
calendar_year_returns = {
    '2010' :factor_returns.loc['31/12/2010'],
    '2011' :factor_returns.loc['31/12/2011'],
    '2012' :factor_returns.loc['31/12/2012'],
    '2013' :factor_returns.loc['31/12/2013'],
    '2014' :factor_returns.loc['31/12/2014'],
    '2015' :factor_returns.loc['31/12/2015'],
    '2016' :factor_returns.loc['31/12/2016'],
    '2017' :factor_returns.loc['31/12/2017'],
    '2018' :factor_returns.loc['31/12/2018'],
    '2019' :factor_returns.loc['31/12/2019'],
    '2020' :factor_returns.loc['31/12/2020'],
    '2021' :factor_returns.loc['31/12/2021'],
    
}
calendar_year_returns = pd.DataFrame(calendar_year_returns)
calendar_year_returns.round(2)
```

```python
corr = factor_returns[['Low Vol','Quality','Size','Momentum','Value','Nifty']].corr()
```

```python
sns.heatmap(corr)
```

```python
active_returns = {
    'Low Vol' : factor_returns['Low Vol'] - factor_returns['Nifty'],
    'Quality' : factor_returns['Quality'] - factor_returns['Nifty'],
    'Size' : factor_returns['Size'] - factor_returns['Nifty'],
    'Momentum' : factor_returns['Momentum'] - factor_returns['Nifty'],
    'Value' : factor_returns['Value'] - factor_returns['Nifty'],
    'High Beta' : factor_returns['High Beta'] - factor_returns['Nifty']
}
active_returns = pd.DataFrame(active_returns)
active_returns.tail()
```

```python
corr_act = active_returns.corr()
corr_act.round(decimals=2)
```

```python
corr_act.style.background_gradient(cmap = 'coolwarm')
```

```python
sns.heatmap(corr_act, cmap= 'RdYlGn',annot=True,fmt="0.2")
```
